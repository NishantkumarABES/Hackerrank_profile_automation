```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("output.csv")
```

```python
data
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
sns.countplot(x ='Problem Solving', data = data, ax=axes[0,0])
axes[0,0].set_title("PROBLEM SOLVING")
axes[0,0].set_xlabel("Number of Stars -->")
axes[0,0].set_ylabel("NUmber of Students -->")

sns.countplot(x ='CPP', data = data,palette='dark', ax=axes[0,1])
axes[0,1].set_title("CPP")
axes[0,1].set_xlabel("Number of Stars -->")
axes[0,1].set_ylabel("NUmber of Students -->")

sns.countplot(x ='Python', data = data,palette='flare', ax=axes[1,0])
axes[1,0].set_title("PYTHON")
axes[1,0].set_xlabel("Number of Stars -->")
axes[1,0].set_ylabel("NUmber of Students -->")

sns.countplot(x ='C language', data = data,palette='crest', ax=axes[1,1])
axes[1,1].set_title("C LANGUAGE")
axes[1,1].set_xlabel("Number of Stars -->")
axes[1,1].set_ylabel("NUmber of Students -->")

plt.show()
```
